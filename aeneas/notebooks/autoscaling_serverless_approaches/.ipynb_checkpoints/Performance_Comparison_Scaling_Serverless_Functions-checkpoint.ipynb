# Serverless platform scaling with rps, keda, cpu

```python
import pandas as pd
```

## Running Time 120s, spawn rate 0.04 rps, for 10 concurrent users

```python
df_k8shpa = pd.read_csv("stats_df_45_120_k8s_0.05.csv")
df_rps = pd.read_csv("stats_df_10_120_rps.csv")
df_kda = pd.read_csv("stats_df_45_120_keda_0.05.csv")
```

```python
from matplotlib import pyplot as plt

plt.plot(df_k8shpa['PT'],df_k8shpa['cdf'],  linestyle='dashed', color='b', marker='.',label='cpu')
plt.plot(df_rps['PT'],df_rps['cdf'],  linestyle='dashdot', color='r', marker='.',label='rps')
plt.plot(df_kda['PT'],df_kda['cdf'],  linestyle='dotted', color='g', marker='.',label='keda')
plt.grid()
plt.xlim([0, 55])
plt.legend(loc='best')
plt.xlabel('PT (Seconds)', fontsize=18)
plt.ylabel('CDF', fontsize=16)
```

## Processing Time


```python
df_k8shpa = pd.read_csv("10_120_k8s_0.05.csv")
df_rps = pd.read_csv("10_120_rps_0.05.csv")
df_kda = pd.read_csv("10_120_keda_0.05.csv")
df = df_k8shpa
df['cpu'] =df_k8shpa['TPT'] 
df['rps']= df_rps['TPT']
df['mql']=df_kda['TPT']
```

```python
bx = df.boxplot(column=['cpu','rps','mql'])  
bx.set_ylabel('Processing Time (Seconds)')
bx.set_xlabel('Scaling Approaches')
```

```python
df_results_keda = pd.read_csv("keda_results/keda_results.csv")
df_results_rps = pd.read_csv("keda_results/rps_results.csv")
df_results_cpu = pd.read_csv("keda_results/cpu_results.csv")
```

```python
from matplotlib import pyplot as plt


#plt.plot(df_results['seconds'],df_results['cpu_requested'],  linestyle='solid', color='g', marker='.',label='keda_cpu_req')
plt.plot(df_results_keda['seconds'],df_results_keda['cpu_used'],   linestyle='dashed', color='g', marker='.',label='keda')
plt.plot(df_results_cpu['seconds'],df_results_cpu['cpu_used'],   linestyle='dotted', color='r', marker='.',label='k8s_hpa')
plt.plot(df_results_rps['seconds'],df_results_rps['cpu_used'],  linestyle='dashdot', color='b', marker='.',label='rps')
plt.grid()

plt.legend(loc='best')
plt.xlabel('Time (Seconds)', fontsize=18)
plt.ylabel('CPU', fontsize=16)
```

```python
from matplotlib import pyplot as plt

#plt.plot(df_results['seconds'],df_results['cpu_requested'],  linestyle='solid', color='g', marker='.',label='keda_cpu_req')
plt.plot(df_results_keda['seconds'],df_results_keda['memory_used'],   linestyle='dashed', color='g', marker='.',label='keda')
plt.plot(df_results_cpu['seconds'],df_results_cpu['memory_used'],   linestyle='dotted', color='r', marker='.',label='k8s_hpa')
plt.plot(df_results_rps['seconds'],df_results_rps['memory_used'],  linestyle='dashdot', color='b', marker='.',label='rps')
plt.grid()

plt.legend(loc='best')
plt.xlabel('Time (Seconds)', fontsize=18)
plt.ylabel('Memory', fontsize=16)
```

```python
df_results_keda['unused']= df_results_keda['cpu_requested']-df_results_keda['cpu_used']
df_results_rps['unused']= df_results_rps['cpu_requested']-df_results_rps['cpu_used']
df_results_cpu['unused']= df_results_cpu['cpu_requested']-df_results_keda['cpu_used']

#plt.plot(df_results['seconds'],df_results['cpu_requested'],  linestyle='solid', color='g', marker='.',label='keda_cpu_req')
plt.plot(df_results_keda['seconds'],df_results_keda['unused'],   linestyle='dashed', color='g', marker='.',label='keda')
plt.plot(df_results_cpu['seconds'],df_results_cpu['unused'],   linestyle='dotted', color='r', marker='.',label='k8s_hpa')
plt.plot(df_results_rps['seconds'],df_results_rps['unused'],  linestyle='dashdot', color='b', marker='.',label='rps')
plt.grid()

plt.legend(loc='best')
plt.xlabel('Time (Seconds)', fontsize=18)
plt.ylabel('CPU Unused', fontsize=16)
```

```python
# importing package
import matplotlib.pyplot as plt

# create data
x = ['keda', 'rps', 'k8s_hpa']
y1 = [df_results_keda['cpu_requested'].mean(),df_results_rps['cpu_requested'].mean(),df_results_cpu['cpu_requested'].mean()]
y2 = [df_results_keda['cpu_used'].mean(),df_results_rps['cpu_used'].mean(),df_results_cpu['cpu_used'].mean()]
y3 = [df_results_keda['unused'].mean(),df_results_rps['unused'].mean(),df_results_cpu['unused'].mean()]

# plot bars in stack manner
plt.bar(x, y2, color='g')
#plt.bar(x, y3, bottom=y2, color='b')
#plt.bar(x, y3, bottom=y1, color='g')
plt.show()

```

```python
fig, ax1 = plt.subplots(figsize=(5, 3))
ax2 = ax1.twinx()


ax2.plot(df_results_cpu['seconds'],df_results_cpu['pod_count'],   linestyle='dashdot', color='r', marker='.',label='pod_count')
#ax1.plot(df_results_cpu['seconds'],df_results_cpu['cpu_requested'],   linestyle='solid', color='g', marker='.',label='pod_count')
ax1.plot(df_results_cpu['seconds'],df_results_cpu['cpu_used'],   linestyle='dashed', color='g', marker='.',label='cpu')
ax2.plot(df_results_cpu['seconds'],df_results_cpu['throughput'],   linestyle='dotted', color='b', marker='.',label='throughput')
#ax1.plot(df_results_rps['seconds'],df_results_rps['cpu_used'],  linestyle='dashdot', color='b', marker='.',label='rps')
plt.xlim([0, 140])
plt.grid()
plt.legend()
#plt.legend(loc='best')
ax1.set_xlabel('Time (Seconds)', fontsize=10)
ax1.set_ylabel('CPU', fontsize=10)
ax2.set_ylabel("# of Pods")
```
